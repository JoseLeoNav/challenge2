### Importación de datos



```python
import pandas as pd
import math
import matplotlib.pyplot as plt


url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
ingreso_total_tienda1= sum(tienda["Precio"])
ingreso_total_tienda2= sum(tienda2["Precio"])
ingreso_total_tienda3= sum(tienda3["Precio"])
ingreso_total_tienda4= sum(tienda4["Precio"])

print(f"Ingreso total de la tienda 1: {ingreso_total_tienda1}")
print(f"Ingreso total de la tienda 2: {ingreso_total_tienda2}")
print(f"Ingreso total de la tienda 3: {ingreso_total_tienda3}")
print(f"Ingreso total de la tienda 4: {ingreso_total_tienda4}")

# Datos para el gráfico
ingresos = [ingreso_total_tienda1, ingreso_total_tienda2, ingreso_total_tienda3, ingreso_total_tienda4]
tiendas = ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"]

# Graficar
plt.plot(tiendas, ingresos, marker='o')
plt.title("Ingresos totales por tienda")
plt.xlabel("Tienda")
plt.ylabel("Ingreso total")
plt.grid(True)
plt.show()
```

# 3. Calificación promedio de la tienda


```python
#Para tienda 1
calificacion_tienda1 = ((sum(tienda["Calificación"]))/len(tienda['Calificación']))
print(calificacion_tienda1)

#Para tienda 2
calificacion_tienda2 = ((sum(tienda2["Calificación"]))/len(tienda2['Calificación']))
print(calificacion_tienda2)

#Para tienda 3
calificacion_tienda3 = ((sum(tienda3["Calificación"]))/len(tienda3['Calificación']))
print(calificacion_tienda3)

#Para tienda 4
calificacion_tienda4 = ((sum(tienda4["Calificación"]))/len(tienda4['Calificación']))
print(calificacion_tienda4)



calificaciones = {
    "Tienda 1": (sum(tienda["Calificación"]) / len(tienda["Calificación"])),
    "Tienda 2": (sum(tienda2["Calificación"]) / len(tienda2["Calificación"])),
    "Tienda 3": (sum(tienda3["Calificación"]) / len(tienda3["Calificación"])),
    "Tienda 4": (sum(tienda4["Calificación"]) / len(tienda4["Calificación"]))
}

df_calificaciones = pd.DataFrame(list(calificaciones.items()), columns=["Tienda", "Calificación Promedio"])


df_calificaciones.plot(kind="bar", x="Tienda", y="Calificación Promedio", legend=False, color='skyblue')
plt.title("Calificación promedio por tienda")
plt.ylabel("Calificación Promedio")
plt.ylim(0, 5)
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

# 4. Productos más y menos vendidos

```python
#Menos vendido en tienda 1
menos_vendido_tienda1 = tienda["Producto"].value_counts().idxmin()
print(menos_vendido_tienda1)

#Menos vendido en tienda 2
menos_vendido_tienda2 = tienda2["Producto"].value_counts().idxmin()
print(menos_vendido_tienda2)

#Menos vendido en tienda 3
menos_vendido_tienda3 = tienda3["Producto"].value_counts().idxmin()
print(menos_vendido_tienda3)

#Menos vendido en tienda 4
menos_vendido_tienda4 = tienda4["Producto"].value_counts().idxmin()
print(menos_vendido_tienda4)


# Datos
tiendas = ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"]
menos_vendidos = [
    tienda["Producto"].value_counts().idxmin(),
    tienda2["Producto"].value_counts().idxmin(),
    tienda3["Producto"].value_counts().idxmin(),
    tienda4["Producto"].value_counts().idxmin()
]


plt.figure(figsize=(8, 4))
bars = plt.bar(tiendas, [1]*4, color="salmon")


for bar, producto in zip(bars, menos_vendidos):
    plt.text(bar.get_x() + bar.get_width()/2, 1.05, producto, ha='center', va='bottom', fontsize=10, rotation=45)


plt.ylim(0, 1.5)
plt.title("Producto menos vendido por tienda")
plt.axis('off')

plt.tight_layout()
plt.show()
```

# 5. Envío promedio por tienda

```python
#Costo tienda 1
costo_envio_tienda1 = (sum(tienda['Costo de envío']))/len(tienda['Costo de envío'])
print(costo_envio_tienda1)

#Costo tienda 2
costo_envio_tienda2 = (sum(tienda2['Costo de envío']))/len(tienda2['Costo de envío'])
print(costo_envio_tienda2)

#Costo tienda 3
costo_envio_tienda3 = (sum(tienda3['Costo de envío']))/len(tienda3['Costo de envío'])
print(costo_envio_tienda3)

#Costo tienda 4
costo_envio_tienda4 = (sum(tienda4['Costo de envío']))/len(tienda4['Costo de envío'])
print(costo_envio_tienda4)



# Datos
tiendas = ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"]
costos_envio = [
    (sum(tienda['Costo de envío'])) / len(tienda['Costo de envío']),
    (sum(tienda2['Costo de envío'])) / len(tienda2['Costo de envío']),
    (sum(tienda3['Costo de envío'])) / len(tienda3['Costo de envío']),
    (sum(tienda4['Costo de envío'])) / len(tienda4['Costo de envío'])
]

# Crear gráfico de puntos
plt.figure(figsize=(8, 5))
plt.scatter(tiendas, costos_envio, color='green', s=100)


plt.title("Costo promedio de envío por tienda")
plt.xlabel("Tiendas")
plt.ylabel("Costo promedio de envío")
plt.grid(True)

for i, costo in enumerate(costos_envio):
    plt.text(i, costo + 0.1, f"{costo:.2f}", ha='center')

plt.tight_layout()
plt.show()
```
